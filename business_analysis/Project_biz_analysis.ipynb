### 4.1 DAU, WAU и MAU. Среднее значение метрик за весь период. Изменения метрик во времени

```python
# создаем функцию построения линейного графика
def line_graf(data,data_col,data_y, color, title, xlabel, ylabel, accuracy, limit):
    # data - датафрейм, в котором хранятся данные для построения графика
    # data_col - название столбца, в котором хранятся значения по оси x
    # data_y - название столбца, в котором хранятся значения по оси y
    # color - цвет линий
    # title - название графика
    # xlabel и ylabel - название осей
    # accuracy - точность, с которой нам следует вывести среднее значение
    # limit - параметр, влияющий на лимиты сверху и снизу
    
    plt.figure(figsize=(15,5)) # задаем размер графика
    sns.set_style('whitegrid',{'axes.facecolor':'1', 'grid.color':'0.9', 'figure.facecolor':'1.0'}) # задаем стиль графика
    sns.lineplot(x=data_col,y=float(data.mean()), data=data, color='red', alpha=.5) # строим среднее значение
    plt.text(data[data_col].min(), float(data.mean())*(limit**0.5), 
             " среднее значение = " + str(data[data_y].mean().round(accuracy)),color='grey', ha='left', fontsize=14) # выводим текстом среднее значение
    
    ma=10 if len(data[data_col])<60 else 0 # задаем размер маркера, если данных немного
    sns.lineplot(x=data_col, y=data_y, data=data, color=color, linewidth=2.2, marker='o', markersize=ma) # строим график
    
    plt.xlim(data[data_col].min(), data[data_col].max()) # задаем ограничения по осям
    plt.ylim(data[data_y].min()/limit,data[data_y].max()*limit)
    
    plt.xticks(alpha=.7, fontsize=14) # задаем параметры отображения осей
    plt.yticks(alpha=.7, fontsize=14)
    plt.xlabel(xlabel, alpha=.7, fontsize=14)
    plt.ylabel(ylabel, alpha=.7, fontsize=14)
    
    plt.title(title, fontsize=22, pad=15, alpha=.7) # задаем название графика
```

```python
# Добавляем столбцы недель, месяцев, годов, дат
visits['month']=visits['start_session'].astype('datetime64[M]')
visits['week']=visits['start_session'].astype('datetime64[W]')
visits['date']=visits['start_session'].dt.date

orders['date']=orders['buy_time'].dt.date
orders['month_order']=orders['date'].astype('datetime64[M]')
orders['week']=orders['date'].astype('datetime64[W]')

costs['month']=costs['dt'].astype('datetime64[M]').astype('str')
```

```python
# группируем данные по дням, неделям, месяцам
mau_total = visits.groupby('month').agg({'uid': 'nunique'}).reset_index()
wau_total = visits.groupby('week').agg({'uid': 'nunique'}).reset_index()
dau_total = visits.groupby('date').agg({'uid': 'nunique'}).reset_index()

# расчет средних значений за весь период
print('среднее число уникальных покупателей в день:', dau_total.uid.mean().round(0))
print('среднее число уникальных покупателей в неделю:', wau_total.uid.mean().round(0))
print('среднее число уникальных покупателей в месяц:', mau_total.uid.mean().round(0))

# построение графиков DAU, WAU, MAU
line_graf(dau_total, 'date', 'uid', 'green', 'DAU', '', 'Кол-во уникальных пользователей',0, 1.1)
line_graf(wau_total.query('week not in "2018-05-31"'), 'week', 'uid', 'green', 'WAU', '', 'Кол-во уникальных пользователей', 0, 1.1)
line_graf(mau_total, 'month', 'uid', 'green', 'MAU', '', 'Кол-во уникальных пользователей', 0, 1.1)
```

**Вывод:**
 
- По графику DAU видно, что число уникальных пользователей, пользующихся сервисом каждый день, растет очень медленно и лишь под конец ноября приходит максимальное число новых пользователей (вероятно, это связано с "Черной Пятницой"). Также наблюдается провал в конце марта, что, скорее всего, связано с проблемами сбора информации или счетчиками на сайте. Среднее число уникальных пользователей в день = 908.

### 4.3 Распределение продолжительности пользовательской сессии за весь период. Выбор средней меры для описания типичной сессии

```python
# создаем функцию построения гистограммы
def histo(data,x,xlabel,ylabel,title,bins,rangen):
    # data - датафрейм, в котором хранятся данные для построения графика
    # x - название столбца, в котором хранятся значения по оси x
    # title - название графика
    # xlabel и ylabel - название осей
    # bins - количество корзин
    # rangen - лимиты по оси х
    
    # задаем размер и стиль графика
    plt.figure(figsize=(12,5)) 
    sns.set_style('whitegrid') 
    
    # строим график
    sns.histplot(data=data, x=x, bins=bins, binrange=rangen, color='green') # построение гистограммы

    # задаём параметры осей
    plt.xlabel(xlabel, alpha=.7, fontsize=14) 
    plt.ylabel(ylabel, alpha=.7, fontsize=14)
    plt.title(title, fontsize=22, pad=15, alpha=.7)

    # построение линий среднего и медианы
    plt.axvline(x=data[x].mean(),color='red', label='mean') if data[x].median()*100> data[x].mean() else print('Среднее значение много больше, чем медиана')
    plt.axvline(x=data[x].median(),color='blue', label='median')
    plt.legend(loc = 'upper right')
```

```python
visits['duration']=(visits['end_session']-visits['start_session'])/np.timedelta64(1,'m') # поскольку секунды в лог файле почти везде отсутствуют, то логичнее длительность сессии измерять в минутах
visits=visits.query('duration>=0') # было несколько визитов с отрицательной продолжительностью. 

histo(visits,'duration', 'Минуты', 'Количество посещений', 'Распределение визитов по длительности', 60, (0,60))

print('Если в качестве средней меры взять медиану, то типичная длительность сессии',visits.duration.median().round(1), 'мин')
print('Если в качестве средней меры взять среднее, то типичная длительность сессии',visits.duration.mean().round(1), 'мин')
```

<div style="border:solid green 4px; padding:10px">

**Вывод:**
    
График скошен вправо, а значит для описания типичной длительности сессии лучше подойдёт медиана.
Медианное значение длительности визита = **5 мин**, а среднее значение = **10.7 мин**.

### 4.4 Расчет Retention Rate с использованием когортного анализа. Изменение метрики во времени

```python
# функция расчета LTV, Retention rate и определения начального числа пользователей в каждой когорте в разрезе устройств и источников
def cohort(df, uid, device, source_id, month_order, revenue, margin_rate):
    
    # df - датафрейм, в котором хранятся данные
    # uid - столбец с уникальными идентификаторами пользователя
    # device - название столбца, в котором хранятся названий устройств
    # source_id - название столбца, в котором хранятся идентификаторы источников
    # month_order - название столбца, в котором хранятся месяца покупки
    # revenue - название столбца, в котором хранятся значение выручки с каждой покупки
    # margin_rate - маржинальность продукта
    
    
    first_month_order=df.groupby(uid)[month_order].min().reset_index()  # находим первый месяц заказа каждого пользователя
    first_month_order.columns=[uid,'first_month_order'] # переименовывем названия столбцов 
    
    temp_1=pd.merge(df,first_month_order, on=uid) # создаем новую таблицу на основе вводного датафрейма. Добавляем туда первый месяц заказа каждого пользователя
    temp_1['age']=((temp_1[month_order]-temp_1['first_month_order'])/np.timedelta64(1, 'M')).round().astype('int64') # определяем время жизни в месяцах для каждого пользователя
    temp_1['purchase_per_user']=temp_1[uid]
    temp_1['first_month_order']=temp_1['first_month_order'].astype('str')

    # группировка по устройствам и источникам зависит от того, какие параметры указываем в функции
    if device==0:
        if source_id==0:
            cohorts=temp_1.groupby(['first_month_order','age']).agg({uid:'nunique',revenue:'sum', 'purchase_per_user':'count'}).reset_index() # определяем когорты
                
            initial_users_count = cohorts[cohorts['age'] == 0][['first_month_order', uid]] # находим начальное кол-во пользователей для каждой когорты
            initial_users_count = initial_users_count.rename(columns={uid:'initial_users'})
            
            retention_rate=cohorts
            retention_rate=retention_rate.merge(initial_users_count, on=['first_month_order']) # к каждой корогте добавляем столбец с начальным кол-вом пользователей
            retention_rate['retention']=retention_rate[uid]/retention_rate['initial_users'] # считаем Retention rate
            
            cohorts=cohorts.groupby(['first_month_order','age']).sum().groupby(level=0).cumsum().reset_index() # считаем число покупок с накоплением по времени жизни когорты
            cohorts=cohorts.merge(initial_users_count, on=['first_month_order']) # к каждой корогте добавляем столбец с начальным кол-вом пользователей

            
        else: 
            cohorts=temp_1.groupby([source_id,'first_month_order','age']).agg({uid:'nunique',revenue:'sum', 'purchase_per_user':'count'}).reset_index() # определяем когорты
            
            initial_users_count = cohorts[cohorts['age'] == 0][[source_id,'first_month_order', uid]] # находим начальное кол-во пользователей для каждой когорты
            initial_users_count = initial_users_count.rename(columns={uid:'initial_users'})
            
            retention_rate=cohorts
            retention_rate=retention_rate.merge(initial_users_count, on=[source_id,'first_month_order']) # к каждой корогте добавляем столбец с начальным кол-вом пользователей
            retention_rate['retention']=retention_rate[uid]/retention_rate['initial_users']# считаем Retention rate
            
            cohorts=cohorts.groupby([source_id,'first_month_order','age']).sum().groupby(level=[0,1]).cumsum().reset_index() # считаем число покупок с накоплением по времени жизни когорты
            cohorts=cohorts.merge(initial_users_count, on=[source_id,'first_month_order']) # к каждой корогте добавляем столбец с начальным кол-вом пользователей
            
    elif source==0:
        cohorts=temp_1.groupby([device,'first_month_order','age']).agg({uid:'nunique',revenue:'sum', 'purchase_per_user':'count'}).reset_index() # определяем когорты
            
        initial_users_count = cohorts[cohorts['age'] == 0][[device,'first_month_order', uid]] # находим начальное кол-во пользователей для каждой когорты
        initial_users_count = initial_users_count.rename(columns={uid:'initial_users'})
        
        retention_rate=cohorts
        retention_rate=retention_rate.merge(initial_users_count, on=[device,'first_month_order']) # к каждой корогте добавляем столбец с начальным кол-вом пользователей
        retention_rate['retention']=retention_rate[uid]/retention_rate['initial_users']# считаем Retention rate
            
        cohorts=cohorts.groupby([device,'first_month_order','age']).sum().groupby(level=[0,1]).cumsum().reset_index() # считаем число покупок с накоплением по времени жизни когорты
        cohorts=cohorts.merge(initial_users_count, on=[device,'first_month_order']) # к каждой корогте добавляем столбец с начальным кол-вом пользователей
        
        
    else: 
        cohorts=temp_1.groupby([device,source,first_month_order,'age']).agg({uid:'nunique',revenue:'sum', 'purchase_per_user':'count'}).reset_index() # определяем когорты
        
        initial_users_count = cohorts[cohorts['age'] == 0][[device,source,'first_month_order', uid]] # находим начальное кол-во пользователей для каждой когорты
        initial_users_count = initial_users_count.rename(columns={uid:'initial_users'})
        
        retention_rate=cohorts
        retention_rate=retention_rate.merge(initial_users_count, on=[device,source_id,'first_month_order']) # к каждой корогте добавляем столбец с начальным кол-вом пользователей
        retention_rate['retention']=retention_rate[uid]/retention_rate['initial_users']# считаем Retention rate
            
        cohorts=cohorts.groupby([device,source,'first_month_order','age']).sum().groupby(level=[0,1,2]).cumsum().reset_index() # считаем число покупок с накоплением по времени жизни когорты
        cohorts=cohorts.merge(initial_users_count, on=[device,source_id,'first_month_order']) # к каждой корогте добавляем столбец с начальным кол-вом пользователей
        
        

    cohorts['margin_profit']=cohorts[revenue]*margin_rate # считаем маржинальную прибыль
    cohorts['ltv'] = cohorts['margin_profit'] / cohorts['initial_users'] # считаем LTV
            
    return cohorts, initial_users_count, retention_rate, temp_1
```

```python
# построение тепловой карты
def graphic(pivot,title,xlabel,ylabel, ac):
    # pivot - входная таблица
    # title - название графика
    # xlabel, ylabel - названия осей
    # ac - точность в ячейках
    
    # задаём стиль и размер графика
    sns.set(style='whitegrid')
    plt.figure(figsize=(14, 5))
    
    # построение тепловой карты
    sns.heatmap(pivot, annot=True, fmt=f'.{ac}', linewidth=0.2, linecolor='green', cmap='YlGn')

    # задаем параметры для осей и их названия
    plt.xticks(alpha=.7, fontsize=14)
    plt.yticks(alpha=.7, fontsize=14)
    plt.xlabel(xlabel, alpha=.7, fontsize=14)
    plt.ylabel(ylabel, alpha=.7, fontsize=14)
    
    plt.title(title, fontsize=22, pad=15, alpha=.7)
```

```python
# определяем retention rate через функцию заданную выше
cohorts,initial_users,retention,base=cohort(orders,'uid',0,0,'month_order','revenue',margin_rate)
retention_pivot=retention[retention['age']>0].pivot_table(index='first_month_order', columns='age', values='retention',aggfunc='mean') # сводная таблица Retention rate для каждой когорты

#строим тепловую карту для Retention rate
graphic(retention_pivot,'Retention rate для разных когорт','возраст когорты','когорта','1%')
```

На графике видно, что Retention rate как правило уменьшается с увеличением жизни когорт (за исключением некоторых когорт, у которых есть локальные максимумы retention rate), что ожидаемо.
    
Стоит отметить, что относительно высокий retention rate в первый месяц не гарантирует высокий retention rate в последующее месяца. Например, если посмотреть на когорты 2017-11 и на 2017-10. Скорее всего, сезонность влияет на этот показатель.

Например, это справедливо для 2017-06 когорты, у которой показатель имеет несколько локальных максимумов. Возможно, рассылка или похожий инструмент вернули пользователей.
    
Вероятно, Retetion rate зависит и от сезонности.

### 6.1 Распределение маркетинговых расходов по источникам за весь период. Изменения метрик во времени. Общая сумма расходов на маркетинг. 

```python
def bar_graf(data, x, y, xlabel, ylabel, title):
    
    # data - датафрейм, в котором хранятся данные для построения графика
    # x - название столбца, в котором хранятся значения по оси x
    # y - название столбца, в котором хранятся значения по оси x
    # title - название графика
    # xlabel и ylabel - название осей
    
    # построение столбчатой диаграммы
    data.plot(x=x,y=y, kind='bar', figsize=(14,5), color='green', legend=None, rot=0, fontsize=14)
    
    # задание параметров для осей
    plt.xlabel(xlabel, alpha=.7, fontsize=14)
    plt.ylabel(ylabel, alpha=.7, fontsize=14)
    plt.title(title, fontsize=20, pad=15, alpha=.7)
    
    plt.xticks(alpha=.7, fontsize=14)
    plt.yticks(alpha=.7, fontsize=14)
```

```python
def multi_graf(data, x, y, hue, xlabel, ylabel, title, par):
    
    # data - датафрейм, в котором хранятся данные для построения графика
    # x - название столбца, в котором хранятся значения по оси x
    # y - название столбца, в котором хранятся значения по оси x
    # title - название графика
    # xlabel и ylabel - название осей
    
    plt.figure(figsize=(14,5))
    sns.set_style('whitegrid',{'axes.facecolor':'1', 'grid.color':'0.9', 'figure.facecolor':'1.0'})
    sns.lineplot(x=x,y=y, data=data, hue=hue, palette="Set2", linewidth=2.2, marker='o', markersize=10)

    if par!=0:
        plt.text(data[x].min(), par*1.04, " среднее значение = " + str(par),color='grey', ha='left', fontsize=14)
        sns.lineplot(x=x,y=par, data=data, color='black', alpha=.5)
    
    plt.xlim(data[x].min(), data[x].max()) # задаем ограничения по осям x

    # задаем параметры осей
    plt.xticks(alpha=.7, fontsize=14, rotation=45)
    plt.yticks(alpha=.7, fontsize=14)
    plt.ylabel(ylabel, alpha=.7, fontsize=14)
    plt.xlabel(xlabel, alpha=.7, fontsize=14)
    
    plt.title(title, fontsize=20, pad=15, alpha=.7)
```
